```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
iris.columns.values
```

```python
# plt.plot('species', 'sepal_length', data=iris, linestyle = 'none', marker ='o')  

plt.scatter(iris['species'], iris['petal_length'])
plt.xlabel('Species')
plt.ylabel('petal length (cm)')
plt.grid() 
```

```python
setosa_data = iris[iris.species == 'setosa']
versicolor_data = iris[iris.species == 'versicolor']
virginica_data = iris[iris.species == 'virginica']

fig, ax = plt.subplots(1,3,figsize=(14,5))

ax[0].hist(setosa_data.petal_length, color='g', label = 'setosa')
ax[1].hist(versicolor_data.petal_length, color='r', label = 'versicolor')
ax[2].hist(virginica_data.petal_length, color='b', label = 'virginica')

ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[0].set_ylabel('Frequency')
ax[1].set_ylabel('Frequency')
ax[2].set_ylabel('Frequency')
ax[0].set_xlabel('petal length (cm)')
ax[1].set_xlabel('petal length (cm)')
ax[2].set_xlabel('petal length (cm)')

plt.show()
```

```python
first_class = titanic_df[titanic_df.Pclass == 1].Survived.value_counts(normalize = True)
second_class = titanic_df[titanic_df.Pclass == 2].Survived.value_counts(normalize = True)
third_class = titanic_df[titanic_df.Pclass == 3].Survived.value_counts(normalize = True)
```

```python
fig, classes = plt.subplots(1,3,figsize=(13,5))
classes[0].pie(first_class, labels = first_class.index.map({0:"Not Survived", 1:"Survived"}), autopct='%1.1f%%',)
classes[1].pie(second_class, labels = second_class.index.map({0:"Not Survived", 1:"Survived"}), autopct='%1.1f%%',)
classes[2].pie(third_class, labels = third_class.index.map({0:"Not Survived", 1:"Survived"}), autopct='%1.1f%%',)
classes[0].set_title('First Class Passengers')
classes[1].set_title('Second Class Passengers')
classes[2].set_title('Third Class Passengers')
plt.show()
```
