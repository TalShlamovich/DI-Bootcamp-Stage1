# GROUPING

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col="user_id" )
```

```python
users
```

#  4. Discover what is the mean age per occupation

```python
users.groupby("occupation").age.mean()
```

#  5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def men1(gender):
    if gender == 'M':
        return 1
    if gender == 'F':
        return 0
    
users['gender_value'] = users['gender'].apply(men1)
men_ratio = users.groupby('occupation')['gender_value'].sum()/users.occupation.value_counts()*100
men_ratio.sort_values()
```

# 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg([min, max])
```

# 7. For each combination of occupation and gender, calculate the mean age


```python
users.groupby(['occupation', 'gender'])['age'].mean()
```

# 8. For each occupation present the percentage of women and men

```python
users.groupby('occupation')['gender'].value_counts(normalize = True)*100
```
